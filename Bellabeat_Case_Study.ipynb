# **`FitBit Fitness Tracker Data`**



### Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

#### Read Data

```python
df = pd.read_csv('dailyActivity_merged.csv')
df.head()
```

```python
df.shape                             # check num of (rows,columns) in data
```

```python
df.columns
```

```python
df['TotalMinutes'] = df['VeryActiveMinutes'] + df['FairlyActiveMinutes'] + df['LightlyActiveMinutes'] + df['SedentaryMinutes']
df                                  # add new column to dataframe
```

```python
df.info()                               # check informations of data
```

```python
df['ActivityDate'] = pd.to_datetime(df['ActivityDate'])
```

```python
df.describe().T                               # check statistics of data
```

#### Data Cleaning

```python
df.isnull().sum()                               # check null values
```

```python
df.duplicated().sum()                           # check duplicated values
```

```python
df.drop(columns =['Id',	'ActivityDate'],axis = 1,inplace = True)
```

#### Data Visualization

```python
df.hist(figsize=(10,10))
plt.show()
```

```python
sns.scatterplot(x='TotalSteps',y='Calories',data=df , c = df.Calories)
plt.title('TotalSteps vs Calories')
plt.show()
```

```python
VeryActiveMinutes = df['VeryActiveMinutes'].sum()
FairlyActiveMinutes = df['FairlyActiveMinutes'].sum()
LightlyActiveMinutes = df['LightlyActiveMinutes'].sum()
SedentaryMinutes = df['SedentaryMinutes'].sum()
minutes = [VeryActiveMinutes, FairlyActiveMinutes, LightlyActiveMinutes, SedentaryMinutes]
labels = ['VeryActiveMinutes', 'FairlyActiveMinutes', 'LightlyActiveMinutes', 'SedentaryMinutes']
plt.pie(minutes, labels=labels, autopct='%1.1f%%')
plt.title('Minutes')
plt.show()
```

```python
sns.pairplot(df)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(), annot=True)
plt.show()
```
